```python
import pandas as pd
import pandas as pd
pd.set_option('display.max_rows', None)
df = pd.read_csv('DataSampahJabar.csv')
df
```

```python
# 2. Dari DataFrame yang telah dibuat, hitunglah total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat untuk tahun tertentu. Tampilkan hasilnya
tahun_dipilih = 2015
total_produksi_sampah = 0

for index, row in df.iterrows():
    if row['tahun'] == tahun_dipilih:
        total_produksi_sampah += row['jumlah_produksi_sampah']

print(f"Total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat pada tahun {tahun_dipilih} adalah sebanyak {total_produksi_sampah:.2f} ton")
```

```python
# 3. Jumlahkan Data Pertahun

datasampah_per_tahun = {}

for index, row in df.iterrows():
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']   
    if tahun not in datasampah_per_tahun:
        datasampah_per_tahun[tahun] = 0
    datasampah_per_tahun[tahun] += jumlah_sampah

print("Jumlah data sampah per tahun adalah sebagai berikut:\n")
for tahun, total in datasampah_per_tahun.items():
    print(f"Tahun {tahun} : {total:2f} ton")
    
df_datasampah_per_tahun = pd.DataFrame(list(datasampah_per_tahun.items()), 
                                     columns=['Tahun', 'Total Produksi Sampah (ton)'])
```

```python
# Export ke CSV
df_datasampah_per_tahun.to_csv('datasampah_per_tahun.csv', index=False)
print("Data diekspor ke CSV.\n")

# Buka file CSV
df_datasampah_per_tahun_csv = pd.read_csv('datasampah_per_tahun.csv')
print("Buka file CSV.")
```

```python
# Export ke Excel
df_datasampah_per_tahun.to_excel('datasampah_per_tahun.xlsx', index=False)
print("Data diekspor ke Excel.\n")

# Buka file Excel
df_datasampah_per_tahun_excel = pd.read_excel('datasampah_per_tahun.xlsx')
print("Buka file excel.")
```

```python
# 4. Jumlahkan data per Kota/Kabupaten per tahun

datasampah_kota_per_tahun = {}

for index, row in df.iterrows():
    kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah_sampah = row['jumlah_produksi_sampah']  
    
    key = (kota, tahun)   
    if key not in datasampah_kota_per_tahun:
        datasampah_kota_per_tahun[key] = 0
    datasampah_kota_per_tahun[key] += jumlah_sampah
print("Jumlah produksi sampah tiap Kota/Kabupaten per tahun adalah sebagai berikut:\n")

for (kota, tahun), total in datasampah_kota_per_tahun.items():
    print(f"{kota}")
    print(f"Tahun {tahun} : {total} ton\n")

df_datasampah_kota_per_tahun = pd.DataFrame(list(datasampah_kota_per_tahun.items()), 
                                     columns=['Tahun', 'Total Produksi Sampah (ton)'])
```

```python
# Export ke CSV
df_datasampah_kota_per_tahun.to_csv('datasampah_kota_per_tahun.csv', index=False)
print("Data diekspor ke CSV.\n")

# Buka file CSV
df_datasampah_kota_per_tahun_csv = pd.read_csv('datasampah_kota_per_tahun.csv')
print("Buka file CSV.")
```

```python
# Export ke Excel
df_datasampah_kota_per_tahun.to_excel('datasampah_kota_per_tahun.xlsx', index=False)
print("Data diekspor ke Excel.\n")

# Buka file Excel
df_datasampah_kota_per_tahun = pd.read_excel('datasampah_kota_per_tahun.xlsx')
print("Buka file excel.")
```

```python
#Jumlahkan total sampah per kabupaten/kota setiap tahun

pd.set_option('display.max_rows', None)

df = pd.read_csv('datasampahjabar.csv')
df

datasampah_kota_per_tahun = df.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_produksi_sampah'].sum().reset_index()

print("Total Sampah per Kabupaten/Kota di Jawa Barat Setiap Tahun")
datasampah_kota_per_tahun
```

```python
#Buatkan rata-rata menggunakan lambda per kabupaten kota
datajumlah = df.groupby(['nama_kabupaten_kota'])['jumlah_produksi_sampah'].sum().reset_index()

datajumlah['Rata-rata'] = df.apply(lambda row: df[df['nama_kabupaten_kota'] == row['nama_kabupaten_kota']]['jumlah_produksi_sampah'].mean(), axis=1)

print("Rata-Rata Sampah per Kabupaten/Kota di Jawa Barat")
datajumlah
```
